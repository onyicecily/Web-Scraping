```python

#conda install gensim==3.8.3
#conda install python-docx
#conda install -c anaconda beautifulsoup4
import requests
from bs4 import BeautifulSoup
from gensim.summarization import summarize
```

```python
url = "https://pau.edu.ng/news/mtn-nigeria-and-pan-atlantic-university-launch-media-innovation-programme-mip/"
```

```python
page = requests.get(url).text
page
```

```python
soup = BeautifulSoup(page)
soup
```

```python
title = soup.find("h1", attrs={"class":"entry-title"}).get_text()
```

```python
p_tags = soup.find_all("p")
p_tags
```

```python
text = [tags.get_text().strip() for tags in p_tags]
text
```

```python
article = " ".join(text)
article
```

```python
import re
```

```python
date = re.search("\d+\s\w+\s\d+",article).group()
date
```

```python
real = re.search("Press Release:.*—-ENDS—",article).group()
real
```

```python
article = re.sub('Press Release: 27 April 2022 MTN', ' ',real).strip()
article
```

```python
summarized = summarize(article,ratio=0.3)
```
